```python
#Calcola la somma cumulativa delle vendite per ogni mese utilizzando il metodo .cumsum()
a['Somma_Cumulativa'] = a['Guadagni'].cumsum()
print(a)
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df_facebook = pd.read_csv(io.BytesIO(uploaded['facebook.csv']))
df_facebook.head()
```

```python
# Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp

df_facebook['status_published'] = pd.to_datetime(df_facebook['status_published'])
```

```python
#Utilizzando il metodo .df.strftime(), formatta le date delle pubblicazioni in un formato specifico (ad esempio, "%Y-%m-%d %H:%M:%S")

df_facebook['status_published_formatted'] = df_facebook['status_published'].dt.strftime("%Y-%m-%d %H:%M:%S")
print(df_facebook[['status_published', 'status_published_formatted']].head())
```

```python
# Ottieniamo informazioni specifiche sulle date delle transazioni
df_facebook['Year'] = df_facebook['status_published'].dt.year
df_facebook['Month'] = df_facebook['status_published'].dt.month
df_facebook['Day'] = df_facebook['status_published'].dt.day
df_facebook['DayOfWeek'] = df_facebook['status_published'].dt.dayofweek
df_facebook['DayOfYear'] = df_facebook['status_published'].dt.dayofyear
print(df_facebook[['status_published', 'Year', 'Month', 'Day', 'DayOfWeek', 'DayOfYear']].head())
```

```python
# Estraiamo solo i post relativi al 2012
df_facebook_2012 = df_facebook[df_facebook['Year'] == 2012]
print(df_facebook_2012.head())
```

```python
# Estraiamo solo i post relativi a maggio 2018
df_facebook_may_2018 = df_facebook[(df_facebook['Year'] == 2018) & (df_facebook['Month'] == 5)]
print(df_facebook_may_2018.head())
```

```python
# Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana
weekend_posts = df_facebook[df_facebook['DayOfWeek'].isin([5, 6])]  # 5 = sabato, 6 = domenica
weekday_posts = df_facebook[~df_facebook['DayOfWeek'].isin([5, 6])]

print(f"Numero di post pubblicati nei weekend: {len(weekend_posts)}")
print(f"Numero di post pubblicati nei giorni feriali: {len(weekday_posts)}")
```

```python
# Troviamo il primo e ultimo post pubblicati in ogni anno
first_posts = df_facebook.loc[df_facebook.groupby('Year')['status_published'].idxmin()]
last_posts = df_facebook.loc[df_facebook.groupby('Year')['status_published'].idxmax()]
print(first_posts[['status_published', 'Year']])
print(last_posts[['status_published', 'Year']])
```

```python
# Utilizzando il metodo .tz_localize(), aggiungiamo un fuso orario alle date delle transazioni
df_facebook['status_published_localized'] = df_facebook['status_published'].dt.tz_localize('Europe/Rome', nonexistent='NaT')
print(df_facebook[['status_published', 'status_published_localized']].head())
```

```python
# Calcoliamo la differenza di tempo tra due transazioni
time_diff = df_facebook['status_published'].diff()
print(time_diff.head())
```

```python
# Quanti tipi di post ci sono? E quanti per ogni tipo?
post_types = df_facebook['status_type'].value_counts()
print(post_types)
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df_automobile = pd.read_csv(io.BytesIO(uploaded['automobile.csv']))
df_automobile.head()
```

```python
# Ci sono valori nulli? Dove? Quanti?
null_values = df_automobile.isnull().sum()
print(null_values)
```

```python
# Quali righe hanno un valore nullo nella colonna num-of-doors?
null_doors = df_automobile[df_automobile['num-of-doors'].isnull()]
print(null_doors)
```

```python
# Esaminando i dati, cerchiamo dei valori da sostituire ai valori nulli nella colonna num-of-doors
# Possiamo vedere che i valori nulli sono presenti in due righe.
# In entrambe le righe, la colonna 'body-style' ha il valore 'hatchback'.
# Possiamo quindi cercare la moda (il valore più frequente) della colonna 'num-of-doors' per i veicoli con 'body-style' = 'hatchback'.

# Troviamo la moda della colonna 'num-of-doors' per i veicoli con 'body-style' = 'hatchback'
mode_doors = df_automobile[df_automobile['body-style'] == 'hatchback']['num-of-doors'].mode()[0]

# Sostituiamo i valori nulli nella colonna 'num-of-doors' con la moda trovata
df_automobile['num-of-doors'].fillna(mode_doors, inplace=True)

# Verifichiamo se ci sono ancora valori nulli nella colonna 'num-of-doors'
print(df_automobile['num-of-doors'].isnull().sum())
```
